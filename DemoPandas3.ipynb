```python
# 선언
from pandas import Series, DataFrame
import pandas as pd

# 시리즈 생성
myCampus = Series([100, 200, 300, 400, 500],
                    index=['2025-12-01', '2025-12-02', '2025-12-03', '2025-12-04', '2025-12-05'])
myCampus
```

```python
sample1 = [["아이폰16", 1500000, 5000] , ["갤럭시S24", 1300000], ["픽셀9", 900000]]
column1 = ["제품명", "가격", "비고"]
df = DataFrame(data=sample1, columns=column1)
df
```

```python
# 딕셔너리 활용
data2 = {'제품명': ["아이폰16", "갤럭시S24", "픽셀9"],
         '가격': [1500000, 1300000, 900000],
         '비고': [5000, None, None]}

data2
df2 = DataFrame(data=data2)
df2
```

```python
# 컬럼 순서 지정
df3 = DataFrame(
        data=data2, 
        columns=['가격', '제품명', '비고'], 
        index=['2025-12-01', '2025-12-02', '2025-12-03'])
df3
```
